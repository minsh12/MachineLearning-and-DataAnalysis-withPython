### **모듈 임포트**
 + 그동안 사용했던 함수들 처럼, 다양한 기능들이 미리 함수로 구현되어 모듈 형태로 제공
 + 대표적으로 추후 과정에서 사용하게 될 아래의 모듈들이 존재

 + requests - HTTP 요청/응답 모듈
 + numpy - 수치해석 모듈 
 + pandas - 데이터 분석 모듈


#### import 
 - import를 사용하여 해당 모듈 전체를 import

```python
import requests
resp = requests.get('http://naver.com')
resp.text
# 네이버 도메인의 html 데이터를 요청한 것
```

```python
import math

print(math.pi)
print(math.cos(100))
```

#### from import 
 - 해당 모듈에서 특정한 타입만 import

```python
from math import pi
from math import cos

pi
```

```python
cos(100)
```

#### \* 임포트
 - 해당 모듈내에 정의된 모든 것을 import
 - 일반적으로 사용이 권장되지 않음

```python
sin(100)
# math.sin(100)으로 사용안했고 & from math import sin도 아직 안해서 오류 발생
```

```python
math.sin(100)
```

```python
from math import *

sin(100)
```

```python
e
```

#### as 
 - 모듈 import 시, alias(별명) 지정가능
 

```python
import math as m

m.exp(3)
```
